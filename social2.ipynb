```python
#EXERCISE 1
#Write a function named get_leaves that takes a graph as an argument, loops through the nodes, and returns a list of nodes with degree 1.
def get_leaves(G):
    for node in G.nodes:
        if G.degree(node)==1:
            print(node)
G = nx.Graph()
G.add_edges_from([
        ('a', 'b'),
        ('a', 'd'),
        ('c', 'd'),
    ])
nx.draw(G, 
        with_labels=True,
        node_size=2000,
        node_color='green',
        font_size=16,
        font_color='white')
get_leaves(G)
#assert set(get_leaves(G)) == {'c', 'b'}
```

```python
#EXERCISE 2
#Write a function max_degree that takes a graph as its argument, and returns a 2-tuple with the name and degree of the node with highest degree.
def max_degree(SG):
    degrees=[val for (node, val) in SG.degree()]
    maxdegree = max(degrees)
    index=degrees.index(maxdegree)
    nodes=[node for (node, val) in SG.degree()]
    node=nodes[index]
    return node,maxdegree
SG = nx.read_adjlist('friends.adjlist')
max_degree(SG)
#assert max_degree(SG) == ('Claire', 4)
```

```python
#EXERCISE 3
def mutual_friends(SG, node_1, node_2):
    mutual=[]
    node1=list(SG.neighbors(node_1))
    node2=list(SG.neighbors(node_2))
    for j in node2:
        if j in node1:
            mutual.append(j)
    return mutual
SG = nx.read_adjlist('friends.adjlist')
print(mutual_friends(SG, 'Alice', 'Claire'))
print(mutual_friends(SG, 'George', 'Bob'))
print(sorted(mutual_friends(SG, 'Claire', 'George')))
```

```python
#directed graph
G=nx.DiGraph()
nodes=['a','b','c','d']
G.add_nodes_from(nodes)
edges=[('a', 'b'),('a', 'c'), ('b', 'c'), ('c', 'd')]
G.add_edges_from(edges)
nx.draw(G, 
        with_labels=True,
        node_size=2000,
        node_color='green',
        font_size=16,
        font_color='white')
```

```python
G.has_edge('a','c')
```

```python
# predecessors ("in-neighbors") and successors ("out-neighbors")
print('Successors of 2:', list(G.successors('a')))
print('Predecessors of 2:', list(G.predecessors('a')))
```

```python
print(G.in_degree('a'))
print(G.out_degree('a'))
```

```python
G.degree('a')
```

```python
print('Successors of 2:', list(G.successors('a')))
#Neighbors contain both Successors and predecessors
print('"Neighbors" of 2:', list(G.neighbors('a')))
```
