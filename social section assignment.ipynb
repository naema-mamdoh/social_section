```python
#looping in List of tuples
student_data=[("sam","01144778899","A-"),
              ("samar","01144778899","B+"),
              ("lily","01144778899","C+"),
              ("tom","01144778899","A"),
              ("hany","01144778899","D")
             ]
for name , phone , grade in student_data:
    if grade.startswith('A'):
        print('Congratulations', name,
              'on getting an', grade)
    elif grade.startswith('B'):
        print('Congratulations', name,
              'on getting an', grade)
    elif grade.startswith('C'):
        print('Congratulations', name,
              'on getting an', grade)
    else:
        print("sorry", name, "you are not passed")
```

```python
#dictionaries
countries = {'egypt':'0020' , 'america':'001' , 'ksa':'00966'}
print ('0020' in countries.values()) #true
# Change an entry that does exist
countries['egypt'] = '0030'
print ('egypt' in countries) #true
print ('0020' in countries) #false
print ('0020' in countries.values()) #false 
# add an entry that doesm't exist
countries['france']='00224'
print(countries)
# Delete an entry that exists
del countries['america']
print(countries)
```

```python
#looping over dictionaries
for key in countries:
    value = countries[key]
    print(key, 'has key', value)
print("###############################################")
for key, value in countries.items():
    print(key, 'has key', value)
```

```python
#List of dictionaries
student_data_list1=[{"name":"sam","phone":"01144778899","grade":"A-"},
              {"name":"samar","phone":"01144778899","grade":"B+"},
              {"name":"lily","phone":"01144778899","grade":"c+"},
              {"name":"tom","phone":"01144778899","grade":"A"},
              {"name":"hany","phone":"01144778899","grade":"D"}
              ]
print(student_data_list1)
print(student_data_list1[2])
print(student_data_list1[2]["name"])
#another way
print("################################################################################################")
student_data_list2 = []
for name, phone, grade in student_data:
    student = {
        'name': name,
        'phone': phone,
        'grade': grade,
    }
    student_data_list2.append(student)
print(student_data_list2)
print(student_data_list2[2])
print(student_data_list2[2]["name"])
```

```python
#Dictionary of dictionaries
student_data_dic1={"sam":{"phone":"01144778899","grade":"A-"},
              "samar":{"phone":"01144778899","grade":"B+"},
              "lily":{"phone":"01144778899","grade":"c+"},
              "tom":{"phone":"01144778899","grade":"A"},
              "hany":{"phone":"01144778899","grade":"D"}
             }
print(student_data_dic1)
print(student_data_dic1["sam"])
print(student_data_dic1["sam"]["grade"])
#another way
print("################################################################################################################")
student_data_dic2 = {}
for name, subject, grade in student_data:
    student = {
        'subject': subject,
        'grade': grade,
    }
    student_data_dic2[name] = student
print(student_data_dic2)
print(student_data_dic2["sam"])
print(student_data_dic2["sam"]["grade"])
```
